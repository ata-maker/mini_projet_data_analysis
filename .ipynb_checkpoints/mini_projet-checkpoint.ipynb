## Importation
Pour mener à bien notre étude et les différents traitement avec Jupyter NoteBook, nous devions importer certaines librairies et des données. 

```python
#import des librairies
import pandas as pd
```

```python
#importation la bd
file = "BF.txt"
```

```python
# Chargement du fichier BF.txt dans un DataFrame Pandas avec les noms de colonnes appropriés
data = pd.read_csv(file, sep='\t', header=None, names=['geonameid', 'name', 'asciiname', 'alternatenames', 'latitude', 'longitude', 'feature_class', 'feature_code', 'country_code', 'cc2', 'admin1_code', 'admin2_code', 'admin3_code', 'admin4_code', 'population', 'elevation', 'dem', 'timezone', 'modification_date'])
```

```python
# Affichage des premières lignes
data.head()
```

```python
#infos
data.info()
```

##  Appliquer les opérations de prétraitement et filtres nécessaires à ce fichier, pour ne garder que les colonnes correspondantes

```python
# Sélection des colonnes nécessaires
data = data[['geonameid', 'name', 'latitude', 'longitude']]
```

```python
# Affichage des premières lignes
data.head()
```

```python
# Renommage des colonnes
data.columns = ['ID', 'location_name', 'lat', 'long']
```

```python
# Affichage des premières lignes du nouveau DataFrame pour vérifier
data.head()
```

## Opérations sur le fichier CSV burkina_location.csv.

```python
# Enregistrement des données dans un fichier CSV
data.to_csv('burkina_location.csv', index=False)
```

```python
# Chargement du fichier burkina_location.csv dans un DataFrame Pandas
location_data = pd.read_csv('burkina_location.csv')
```

```python
# Affichage des premières lignes
location_data.head()
```

```python
# Extraction des données contenant le nom 'gounghin'
data_gounghin = location_data[location_data['location_name'].str.contains('gounghin', case=False)]
```

```python
# Affichage des premières lignes
data_gounghin
```

```python
# Enregistrement des données dans le fichier gounghin.csv
data_gounghin.to_csv('gounghin.csv', index=False)
```

```python
# Filtrage des données pour les noms de lieux commençant par les lettres 'A' à 'P'
data_subset = location_data[location_data['location_name'].str[0].str.upper().between('A', 'P')]
```

```python
# Affichage des premières lignes
data_subset.head()
```

```python
# Recuperation de l'index de la ligne avec la latitude minimale
min_lat_index = data_subset['lat'].idxmin()

# Recuperation de l'index de la ligne avec la longitude minimale
min_lon_index = data_subset['long'].idxmin()

# Recuperation des lignes correspondantes
min_lat_line = data_subset.loc[min_lat_index]
min_lon_line = data_subset.loc[min_lon_index]

# Affichage du résultat
print("Lieu avec la latitude minimale:")
print(min_lat_line[['location_name', 'lat', 'long']])
print("\nLieu avec la longitude minimale:")
print(min_lon_line[['location_name', 'lat', 'long']])
```

```python
# les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)
data_coordinates_filter = data_subset[(data_subset['lat'] >= 11) & (data_subset['long'] <= 0.5)]

# Affichage des résultats
print("Lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5):")
print(data_coordinates_filter[['location_name', 'lat', 'long']])
```

## Sorties Excel
J'ai opté d'utiliser ExcelWriter de pandas.
Le code crée un fichier Excel nommé "mini_projet.xlsx" et écrit deux ensembles de données sur différentes feuilles de ce fichier. Le premier ensemble de données contient des informations sur les lieux contenant le nom "gounghin", et le deuxième ensemble de données contient des informations sur les lieux dont les noms commencent par les lettres 'A' à 'P'. Un message de confirmation est ensuite affiché pour indiquer que les données ont été enregistrées avec succès dans le fichier Excel.

```python
# Création d'un objet ExcelWriter pour écrire dans un fichier Excel
with pd.ExcelWriter('mini_projet.xlsx') as writer:
    # Création une feuille pour les données contenant le nom 'gounghin'
    if not data_gounghin.empty:  # Vérifirication si des données ont été extraites
        data_gounghin.to_excel(writer, sheet_name='gounghin', index=False)

    # Création une feuille pour les données dont les noms de lieux commencent par les lettres 'A' à 'P'
    if not data_subset.empty:  # Vérifirication si des données ont été extraites
        data_subset.to_excel(writer, sheet_name='A_to_P', index=False)

#  message de confirmation
print("Les données ont été enregistrées dans le fichier mini_projet.xlsx avec succès.")
```
